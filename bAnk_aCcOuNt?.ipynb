```python
class bank:
  def __init__(self,name,age):
    self.name=name
    self.age=age
    self.balance=0

  def showbalance(self):
    print(f"The balance for {self.name} is {self.balance}")

  def withdraw(self,money):
    if money>self.balance:
      print("Not enought balance")
    else:
      self.balance=self.balance-money
      print(f'The new balance is {self.balance}')


  def deposit(self,money):
    self.balance=self.balance+money
    print(f'The new balance is {self.balance}')
```

```python
arika=bank('Arika', 12)
```

```python
arika.showbalance()
```

```python
arika.withdraw(3002)
```

```python
arika.deposit(1)
```
