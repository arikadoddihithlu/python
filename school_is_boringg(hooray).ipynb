```python
class student:
  def __init__(self,name,dddd,grade):
    self.name=name
    self.DOB=dddd
    self.grade=grade

  def show_grade(self):
    print(f"{self.name}'s current grade is {self.grade}.")

  def show_age(self):
    age=2026-int(self.DOB)
    print(f"{self.name}'s age right now is {age}.")
```

```python
arika=student('Arika',2014,'7')
```

```python
akira.show_age()
```

```python
class subject:
  def __init__(self, parm_name,parm_materials,parm_teacher,parm_classroom):
    self.name=parm_name
    self.materials=parm_materials
    self.teacher=parm_teacher
    self.classroom=parm_classroom

  def show_materials(self):
    print(f"The matirals u need is {self.materials}")

  def change_teacher(self,new_teacher):
    self.teacher=new_teacher

  def show_teacher(self):
    print(f"Hello! Your teacher will be {self.teacher}")
```

```python
science=subject('Science','Notebook','Miss Taylor Swift','F267')
```

```python
science.show_teacher()
```
